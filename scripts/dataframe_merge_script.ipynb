```python
import re

date_ptr = re.compile(r'1\d{5}')
```

```python
def bbasak_df(keyword):
    df = pd.read_csv('./dataframes/bbasak_{}.csv'.format(keyword),
                     names=['id', 'date', 'title', 'content', 'view', 'c_time'])
    df['product'] = keyword
    df_cols = ['id', 'date', 'title', 'content', 'product']
    df = df[df_cols]
    
    df['date'] = df['date'].to_string()
    df['date'] = df['date'].str.replace("-", "")
    for i in range(0, len(df)):
        df['date'][i] = date_ptr.search(df['date'][i]).group(0)

    return df

def bbasak_concat(*args):
    keywords = []
    
    for keyword in args:
        keywords.append(keyword)
        
    df = pd.concat(keywords)
    df.index = range(len(df))
    df['ref'] = 'bbasak'
    return df

bassak_g5 = bbasak_df('g5')
bassak_g6 = bbasak_df('g6')
bassak_s7 = bbasak_df('s7')

bbasak_df = bbasak_concat(bassak_g5, bassak_g6, bassak_s7)
bbasak_df.head()
```

```python
def dcinside_df(keyword):
    df = pd.read_csv('./dataframes/dcinside_{}.csv'.format(keyword),
                     names=['id', 'date', 'title', 'content', 'view', 'c_time'])
    df['product'] = keyword
    df_cols = ['id', 'date', 'title', 'content', 'product']
    df = df[df_cols]
    
    df['date'] = df['date'].to_string()
    df['date'] = df['date'].str.replace("-", "")
    for i in range(0, len(df)):
        df['date'][i] = date_ptr.search(df['date'][i]).group(0)
        
    return df

def dcinside_concat(*args):
    keywords = []
    
    for keyword in args:
        keywords.append(keyword)
        
    df = pd.concat(keywords)
    df.index = range(len(df))
    df['ref'] = 'dcinside'
    return df

dcinside_g5 = dcinside_df('g5')
dcinside_g6 = dcinside_df('g6')
dcinside_s7 = dcinside_df('s7')

dcinside_df = dcinside_concat(dcinside_g5, dcinside_g6, dcinside_s7)
dcinside_df.head()
```

```python
def ppomppu_df(keyword):
    df = pd.read_csv('./dataframes/ppomppu_{}.csv'.format(keyword),
                     usecols=[0, 1, 2, 3, 4, 5],
                     names=['id', 'date', 'view', 'title', 'content', 'c_time'])
    df['product'] = keyword
    df_cols = ['id', 'date', 'title', 'content', 'product']
    df = df[df_cols]
    
    df['date'] = df['date'].to_string()
    df['date'] = df['date'].str.replace("-", "")    
    for i in range(0, len(df)):
        df['date'][i] = date_ptr.search(df['date'][i]).group(0)
        
    return df

def ppomppu_concat(*args):
    keywords = []
    
    for keyword in args:
        keywords.append(keyword)
        
    df = pd.concat(keywords)
    df.index = range(len(df))
    df['ref'] = 'ppomppu'
    return df

ppomppu_g5 = ppomppu_df('g5')
ppomppu_g6 = ppomppu_df('g6')
ppomppu_s7 = ppomppu_df('s7')

ppomppu_df = ppomppu_concat(ppomppu_g5, ppomppu_g6, ppomppu_s7)
ppomppu_df.tail()
```

```python
def ilbe_df(keyword):
    df = pd.read_csv('./dataframes/ilbe_{}.csv'.format(keyword))
    df_cols = ['id', 'date', 'title', 'content', 'product']
    df = df[df_cols]
    df['content'] = df['content'].str.replace("이 게시물을...", "")
    
    df['date'] = df['date'].to_string()
    df['date'] = df['date'].str.replace("-", "")    
    for i in range(0, len(df)):
        df['date'][i] = date_ptr.search(df['date'][i]).group(0)
        
    return df

def ilbe_concat(*args):
    keywords = []
    
    for keyword in args:
        keywords.append(keyword)
        
    df = pd.concat(keywords)
    df.index = range(len(df))
    df['ref'] = 'ilbe'
    return df

ilbe_g5 = ilbe_df('g5')
ilbe_g6 = ilbe_df('g6')
ilbe_s7 = ilbe_df('s7')

ilbe_df = ilbe_concat(ilbe_g5, ilbe_g6, ilbe_s7)
ilbe_df.tail()
```

```python
phone_df = pd.concat([bbasak_df, dcinside_df, ppomppu_df, ilbe_df])
phone_df.index = range(len(phone_df))
phone_df.tail()
```

```python
phone_frame = phone_df.drop_duplicates(['id'])
phone_frame = phone_df.drop_duplicates(['title', 'content'])
phone_frame = phone_frame.replace(np.nan, 'empty cell')
phone_frame.index = range(len(phone_frame))
phone_frame.head()
```

```python
def filtering(df, keyword):
    for col in ['title', 'content']:
        df[col] = df[col].str.lower()
        df[col] = df[col].str.replace("\\", "")
        df = df[~df[col].str.contains('|'.join(keyword))]
    df.index = range(len(df))
    return(df)
```

```python
filtering_keywords = ['empty cell', '번두목', 'acer', 'amazon', 'hp', 'ebay']
```

```python
phone_frame = filtering(phone_frame, filtering_keywords)
phone_frame.tail()
```
